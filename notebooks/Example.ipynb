# Example Verification Notebook (Binder)
Verifies Python, CPU, RAM, network, and the FastAPI proxy at `/proxy/50080/`. Free-only, no telemetry.

```python
import os, sys, platform, json, time, socket
import psutil, requests
print('Python:', sys.version)
print('Platform:', platform.platform())
print('CPU count:', psutil.cpu_count())
vm = psutil.virtual_memory()
print('RAM total (GB):', round(vm.total/1024**3, 2))
print('Hostname:', socket.gethostname())
```

```python
# Quick CPU benchmark (pure Python)
def fib(n):
    a,b=0,1
    for _ in range(n):
        a,b=b,a+b
    return a
t0=time.time(); x=fib(200000); t1=time.time()
print('FIB(200k) computed len:', len(str(x)), 'time:', round(t1-t0,3),'s')
```

```python
# Verify the FastAPI proxy is reachable at /proxy/50080/
# On mybinder.org this should route to the uvicorn FastAPI started by `start`
URL = 'http://127.0.0.1:50080/'
try:
    r = requests.get(URL, timeout=10)
    print('HTTP', r.status_code)
    print('Body:', r.text[:200])
except Exception as e:
    print('Proxy check failed:', e)
```
