# Stress Test (Binder)
Runs multi-core CPU load and memory allocation test within safe limits.

```python
import concurrent.futures, time, os, psutil
def work(n):
    s=0
    for i in range(5_000_00):
        s += (i*n) % 97
    return s
t0=time.time()
cores = min( max(1, psutil.cpu_count() or 1), 4)
print('Using cores:', cores)
with concurrent.futures.ThreadPoolExecutor(max_workers=cores) as ex:
    futs=[ex.submit(work, i+1) for i in range(cores)]
    res=[f.result() for f in futs]
print('Results sample:', res[:3])
print('CPU test time:', round(time.time()-t0,2),'s')
# Memory test (safe): allocate ~64MB and free
t0=time.time(); buf=bytearray(64*1024*1024);
for i in range(0,len(buf),4096): buf[i]=i%256
dt=time.time()-t0
del buf
print('Memory test 64MB OK in', round(dt,2),'s')
```
